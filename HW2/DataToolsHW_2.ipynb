```python
camera = pd.read_csv('https://raw.githubusercontent.com/CUNY-CISC-3225/datasets/main/cameras.csv')

print(camera.head())
```

```python
# What is the average cost of all the cameras?
avg_cost=camera['Price'].mean()
print(avg_cost)
```

```python
#What is the brand and model name of the most expensive camera? There may be more than one.
max_price=camera['Price'].max()
expensive_cameras=camera[camera['Price']==max_price]
print(expensive_cameras[['Brand','Model']])
```

```python
#Which are more common, below-average-cost cameras or above-average-cost cameras?
above_average= camera[camera['Price']>avg_cost].shape[0] #to count the number of rows
below_average=camera[camera['Price']<avg_cost].shape[0]
print(above_average)
print(below_average)
#Below average costing cameras are more common
```

```python
#Divide the DataFrame into two new DataFrames: one with cameras released on or before 2002, and one with cameras released after 2002
cameras_before2002 = camera[camera['Release date'] <= 2002]
cameras_after2002 = camera[camera['Release date'] > 2002]
print(cameras_before2002)
print(cameras_after2002)
```

```python
# On average, did any camera characteristics improve over time? Which ones?
before_2002means = cameras_before2002.mean(numeric_only=True)
after_2002means = cameras_after2002.mean(numeric_only=True)

improved_characteristics = before_2002means[before_2002means < after_2002means].index
print(improved_characteristics)
# Max Resolution, Low Resolution, Effective pixels, Zoom tele(T) and storage included improved over time.
```

```python
#On average, did any camera characteristics stay about the same? Which ones?
unchanged_characteristics = before_2002means[before_2002means.round(4) == after_2002means.round(4)].index
print("Unchanged characteristics:")
#None of the camera characteristics remained the same on average.
```

```python
#On average, did any camera characteristics get worse over time? Which ones?
worsened_characteristics = before_2002means[before_2002means > after_2002means].index
print(worsened_characteristics)
#Zoom wide, Normal Focus range, Macro focus range, Weight, Dimensions and price got worse over time.
```

```python
#Which camera(s) have the most favorable price-to-effective-pixels ratio? There may be more than one.
#Create a new column in the DataFrame to help you answer this question.
camera['Price_Per_Effective_Pixel'] = camera['Price'] / camera['Effective pixels']
best_ratio = camera[camera['Price_Per_Effective_Pixel'] == camera['Price_Per_Effective_Pixel'].min()]
print(best_ratio[['Brand', 'Model']])
```
